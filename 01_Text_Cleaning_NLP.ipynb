# Text Cleaning

```python
sample_text = "Url https://youtube.com email abc3@gmail.com <h1>Removing Tags </h1>"
sample_text
```

```python
import re
def text_clean(text):
  url = re.compile(r'https?://\S+|www\.\S+')
  text = url.sub('', text)

  email = re.compile(r'[A-Za-z0-9._%+-]+@[\w]+\.[\w]+')
  text = email.sub('', text)

  tags = re.compile(r'<.*?>')
  text = tags.sub('', text)
  return text
text_clean(sample_text)
```

# Remove Punctuation

```python
import pandas as pd

data = {
    "ID": [1, 2, 3, 4, 5],
    "Review": [
        "Wow!!! This movie was AMAZING!!!  Totally loved it!!! #BestMovieEver!!!",
        "Ugh... what a waste of time???  Soooo boring!!!",
        "Hmm... it was okay-ish???  Not great, not bad... just \"meh\"!!!",
        "LOL!!! The acting was sooo bad!!!??? I can't even!!!  #Cringe",
        "Brilliant!!! Loved the story-line, characters & music!!! Must-watch!!!"
    ],
    "Sentiment": ["Positive", "Negative", "Neutral", "Negative", "Positive"]
}

df = pd.DataFrame(data)
df
```

```python
import string
exclude = string.punctuation
```

```python
def remove_pun(text):
  return text.translate(str.maketrans(' ',' ', exclude))
df['Review'] = df['Review'].apply(remove_pun)
df
```
