# Text Representation (Classical Methods)

### BOG (Bag of Words)

Bag of Words (BoW) is a simple and commonly used method for representing text data in natural language processing (NLP). It involves creating a "bag" of all the unique words in a document or a collection of documents, and then representing each document as a vector that counts the frequency of each word in the bag. The order of the words is not considered, hence the name "bag of words". This method is often used for tasks like text classification and sentiment analysis.

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
data = {
    "text": [
        "The quick brown fox jumps over the lazy dog.",
        "Artificial intelligence is transforming the way we work and communicate.",
        "Despite the heavy rain, the outdoor concert continued as scheduled.",
        "Have you ever wondered why the sky appears blue during the day?",
        "The recipe calls for exactly two cups of flour and a pinch of salt."
    ]
}

df = pd.DataFrame(data)
```

```python
cv = CountVectorizer()
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
```

```python
cv.transform(['Artificial intelligence intelligence is and two cup of floor']).toarray()
```

# N_grams

Ngrams are a contiguous sequence of n items from a given sample of text or speech. They are used in natural language processing (NLP) to capture the context and relationships between words in a text. The "n" in n-grams represents the number of items in the sequence. For example, if n=2, we have bigrams, which are pairs of consecutive words; if n=3, we have trigrams, which are triplets of consecutive words.

```python
from sklearn.feature_extraction.text import CountVectorizer

docs = ["I love NLP", "I love AI"]


vectorizer = CountVectorizer(ngram_range=(3,3), stop_words=None, token_pattern=r'(?u)\b\w+\b') # Allow single-character words
X = vectorizer.fit_transform(docs)
print(vectorizer.vocabulary_)
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

### TF-IDF

TD-IDF (Term Frequency-Inverse Document Frequency) is a statistical measure used to evaluate the importance of a word in a document relative to a collection of documents (corpus). It is calculated by multiplying two metrics: Term Frequency (TF) and Inverse Document Frequency (IDF).

```python
from sklearn.feature_extraction.text import TfidfVectorizer

docs = ["I love NLP", "I love AI", "NLP is fun"]

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(docs)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```
