```python
my_vocab = [
    "mango", "computer", "desk", "sedan",
    "apple", "tab", "chair", "coupe",
    "banana", "ipad", "sofa", "motorcycle",
    "orange", "laptop", "table", "truck"
]
```

```python
glove_model['mango']
```

```python
words=[]
vectors=[]
for word in my_vocab:
  words.append(word)
  vectors.append(glove_model[word])
```

```python
corp = zip(words, vectors)
model = TSNE(n_components=2, perplexity=4, init='pca', random_state=42)
fit_model = model.fit_transform(np.array(vectors))
```

```python
x=[]
y=[]
for i in fit_model:
  x.append(i[0])
  y.append(i[1])
```

```python
plt.figure(figsize=(10,8))
for i in range(len(words)):
  plt.scatter(x[i], y[i])
  plt.annotate(words[i], xy=(x[i], y[i]))
```
