# Word2Vec

**word2vec** is a popular technique in natural language processing (NLP) for learning vector representations of words. It was introduced by Mikolov et al. in 2013 and has since become a fundamental tool for various NLP tasks.
**Why word2vec is better than td-idf?**
Word2Vec and TF-IDF are both popular techniques for representing text data, but they serve different purposes and have distinct advantages and disadvantages. Here are some reasons why Word2Vec is often considered better than TF-IDF:
1. **Contextual Understanding**: Word2Vec captures the context of words in a sentence, allowing it to understand the meaning of words based on their surrounding words. TF-IDF, on the other hand, only considers the frequency of words in a document and does not capture any contextual information.
2. **Semantic Relationships**: Word2Vec can capture semantic relationships between words, such as synonyms and analogies. For example, it can understand that "king" is to "queen" as "man" is to "woman". TF-IDF does not capture these relationships and treats each word as an independent entity.
3. **Dimensionality Reduction**: Word2Vec generates dense vector representations of words, which can be more efficient for machine learning models. TF-IDF produces sparse vectors, which can be high-dimensional and may require additional techniques for dimensionality reduction.
4. **Generalization**: Word2Vec can generalize better to unseen data because it learns the underlying structure of the language. TF-IDF is more specific to the training data and may not perform well on new, unseen documents.
5. **Performance**: In many NLP tasks, Word2Vec has been shown to outperform TF-IDF, especially in tasks that require understanding the meaning of words and their relationships. For example, in tasks like sentiment analysis, word similarity, and machine translation, Word2Vec often yields better results than TF-IDF.
Here is a simple example of how to use Word2Vec in Python using the Gensim library:


```python
import pandas as pd
import numpy as np
import gensim
import nltk
from gensim.models import Word2Vec
```

```python
df = pd.read_csv('/content/Twitter_Data.csv')
df.shape
```

```python
df.head(2)
```

```python
Review_Text = df.clean_text.fillna('').apply(gensim.utils.simple_preprocess)
Review_Text.sample(10)
```

```python
model = Word2Vec(
    sentences=Review_Text,
    window=5,
    min_count=5,
    vector_size=200,
    sg=0
)
```

```python
model.build_vocab(Review_Text, progress_per=500)
```

```python
model.train(Review_Text, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
model.wv.most_similar('nonsense')
```

```python
#How often a word appeared during training
model.wv.get_vecattr("kin", "count")
```

```python
model.wv.most_similar(positive=["sir", "mam"], negative=["human"])
```

```python
model.wv.most_similar(positive=["king", "woman"], negative=["man"])
```
