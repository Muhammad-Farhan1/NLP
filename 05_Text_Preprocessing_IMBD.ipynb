# Text Preprocessing

```python
# Removing HTML tags
def Remove_tags(text):
  tags = re.compile(r'<[^>]+>')
  text = tags.sub('', text)
  return text
Remove_tags(df['review'][1])
```

```python
# Lowercasing
def Lowercasing(text):
  lower_text = text.lower()
  return lower_text
Lowercasing(df['review'][1])
```

```python
# Remove punctuation
def Remove_Punct(text):
  clean_text = re.compile(r'[^\w\s]')
  text = clean_text.sub('', text)
  return text
Remove_Punct(df['review'][1])
```

```python
# Remove stopwords
nltk.download('stopwords')
stopword = stopwords.words('english')

def Remove_Stopwords(text):
  new_text = []

  for word in text.split():
    stopword = set(stopwords.words('english'))
    new_text.append(word)
  return " ".join(new_text)
Remove_Stopwords(df['review'][1])
```

```python
# Tokenization
def tokenize(text):
  final_text = word_tokenize(text)
  return final_text
tokenize(df['review'][1])
```

```python
# Stemming
ps = PorterStemmer()

def stem_word(text10):
  return ' '.join([ps.stem(word) for word in text10.split()])
stem_word(df['review'][0])
```

```python
wnl = WordNetLemmatizer()
import nltk
nltk.download('averaged_perceptron_tagger_eng') # Download the missing resource
nltk.download('wordnet') # Download the missing wordnet resource
from nltk.corpus import wordnet
def lem(text):
    def get_wordnet_pos(tag):
        if tag.startswith('J'):
            return wordnet.ADJ
        elif tag.startswith('V'):
            return wordnet.VERB
        elif tag.startswith('N'):
            return wordnet.NOUN
        elif tag.startswith('R'):
            return wordnet.ADV
        else:
            return wordnet.NOUN

    words = nltk.word_tokenize(text)
    words = [word for word in words if word.isalnum()]
    pos_tags = nltk.pos_tag(words)

    lemmatized = [wnl.lemmatize(word, get_wordnet_pos(tag)) for word, tag in pos_tags]

    return " ".join(lemmatized)
lem(df['review'][2])
```

```python
port_stem = PorterStemmer()
def stemming(content):
    stemmed_content = re.sub('[^a-zA-Z]', ' ', content)
    stemmed_content = stemmed_content.lower()
    stemmed_content = stemmed_content.split()
    stemmed_content = [port_stem.stem(words) for words in stemmed_content if words not in stopwords.words('english')]
    stemmed_content = ' '.join(stemmed_content)
    return stemmed_content
```
